#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Необходимо построить модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Мне нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

Прежде всего, импортирую нужные библиотеки.

## Обучение

Прежде всего, напишу функцию для создания признаков. Признаки:
1. День недели;
2. Час;
3. Лаг (отставание, отстающие значения);
4. Скользящее среднее.

```python
def make_features(data, max_lag, rolling_mean_size):
    data['hour'] = data.index.hour
    data['dayofweek'] = data.index.dayofweek

    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    data = data.dropna()
```

Применю функцию с лагов в сутки (24 часа), со скользящим средним - 50.

```python
make_features(forecast, 24, 168)
```

```python
# Код ревьюера
from statsmodels.graphics import tsaplots

taxi = pd.read_csv("/datasets/taxi.csv", index_col=[0], parse_dates=[0]).resample('1H').sum()
ts = taxi['num_orders']

fig = tsaplots.plot_acf(ts,lags= 190)
fig.set_size_inches(18.5, 6)
plt.ylim(-.1, 1.1)
plt.show()
```

Избавлюсь от пропусков

```python
forecast = forecast.dropna()
```

Поделю данные данные на обучающую и тестовую выборки. Тестовая - 10% согласно условию.

```python
train, test = train_test_split(forecast, shuffle=False, test_size=0.1)

features_train = train.drop(['num_orders'], axis=1)
target_train = train['num_orders']
features_test = test.drop(['num_orders'], axis=1)
target_test = test['num_orders']
```

Проверю размерность выборок.

```python
print(features_train.shape)
print(features_test.shape)
print(target_train.shape)
print(target_test.shape)
```

Задам функцию для корня из среднеквадратической ошибки.

```python
def RMSE(target, predict):
    return (mean_squared_error(target, predict))**0.5

RMSE_sc = make_scorer(RMSE, greater_is_better=False)
```

### Обучение модели линейной регрессии

Имею дело с временным рядом, значит не могу применять кросс-валидацию  для подбора гиперпараметров моделей с помощью GridSearchCV. Использую TimeSeriesSplit().

```python
reg_cv = TimeSeriesSplit(n_splits=3).split(features_train)
```

Называю и обучаю модель

```python
%%time
regression = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=True)
regression.fit(features_train, target_train)
```

Считаю корень из среднеквадрачической ошибки.

```python
%%time
regression_cv_score = pd.Series(cross_val_score(regression,features_train,
                                                target_train,
                                                scoring=RMSE_sc,cv=reg_cv)).mean()*(-1)
print('Среднее качество модели линейной регрессии на кросс-валидации:', regression_cv_score)
```

Итак, RMSE для линейной регрессии 28,09.

### Обучение модели случайного леса

Действую аналогично предыдущей модели.

```python
forest_cv = TimeSeriesSplit(n_splits=3).split(features_train)
```

Называю модель и словарь с набором гиперпараметров.

```python
%%time
forest = RandomForestRegressor()

forest_parametrs = { 'n_estimators': range(2, 34, 2),
                     'max_depth': range(1, 15, 1),
                     'min_samples_leaf': range(1,5),
                     'min_samples_split': range(2,10,2) }
```

Применяю GridSearchCV()

```python
%%time
forest_grid = GridSearchCV(forest, forest_parametrs, scoring=RMSE_sc, cv=forest_cv)
forest_grid.fit(features_train, target_train)
```

Определяю качество модели.

```python
%%time
forest_params = forest_grid.best_params_
forest_score = forest_grid.best_score_*(-1)
print(forest_params)
print('Среднее качество модели случайного леса на кросс-валидации:',forest_score)
```

### Обучение модели категориального бустинга

```python
cb_cv = TimeSeriesSplit(n_splits=3).split(features_train)
```

Называю модель и словарь параметров.

```python
%%time
cb = CatBoostRegressor()
cb_parametrs = {'learning_rate': [0.03, 0.1],
                'depth': [4, 6, 10]
               }
```

Применяю GridSearchCV()

```python
%%time
cb_grid = GridSearchCV(cb, cb_parametrs, scoring=RMSE_sc, cv=cb_cv)
cb_grid.fit(features_train, target_train, verbose=0)
```

Определяю качество модели.

```python
%%time
cb_params = cb_grid.best_params_
cb_score = cb_grid.best_score_ * (-1)
print(cb_params)
print(cb_score)
```
