### Проверка правильности расчёта эффективности обогащения

Вычислю эффективность обогащения на обучающей выборке для признака rougher.output.recovery. Определю среднюю абсолютную ошибку между расчётами и значением признака.

C — доля золота в концентрате после флотации/очистки;
F — доля золота в сырье/концентрате до флотации/очистки;
T — доля золота в отвальных хвостах после флотации/очистки

```python
c = data_train['rougher.output.concentrate_au']
f= data_train['rougher.input.feed_au']
t = data_train['rougher.output.tail_au']
input_recovery = data_train['rougher.output.recovery']
calculated_recovery = ((c * (f - t)) / (f * (c - t))) * 100
print('Эффективность обогащения чернового концентрата', calculated_recovery)
print('Абсолютная средняя ошибка =', mean_absolute_error(input_recovery, calculated_recovery))
```

Ошибка получилась небольшая. Значения посчитаны верно.

### Функция для вычисления итоговой sMAPE

Построю функцию sMape для вычисления одного этапа.

```python
def smape_function (target, predic):

    smape = abs(target-predic)/((abs(target)+abs(predic))/2)*100
    smape = smape.fillna(value=0)

    smape = np.mean(smape)
    return smape
```

Построю итоговую sMape функцию при входных данных от сырья и финала.

```python
def total_smape_function(target_roug, predic_rough, target_final, predic_final):

    smape_roug = smape_function(target_roug, predic_rough)

    smape_final = smape_function(target_final, predic_final)

    total_smape = 0.25 * smape_rough + 0.75 * smape_final

    return total_smape
```

И, наконец, итоговая sMape при входных sMape rough и sMape final.

```python
def final_smape(smape_roug, smape_final):

    final_smape = 0.25 * smape_roug + 0.75 * smape_final

    return final_smape
```

#### Проверка лучшей модели на тестовой выборке

Наименьшая ошибка у модели случайного леса. Её и проверим на тестовой выборке.

```python
def forest_model(features_train, target_train, features_test, depth, estimators):

    model_forest = RandomForestRegressor(random_state=12345, max_depth = depth, n_estimators=estimators)

    model_forest.fit(features_train, target_train)

    pred_forest_test = model_forest.predict(features_test)

    return pred_forest_test
```

Прогноз эффективности для этапа флотации.

```python
pred_test_rough = forest_model(features_train, target_train['rougher.output.recovery'], features_test, 5, 130)
```

Прогноз эффективности для финального концентрата.

```python
pred_forest_final = forest_model(features_train, target_train['final.output.recovery'], features_test, 3, 50)
```

Расчёт sMAPE на тестовых данных.

```python
smape_test = total_smape_function(target_test['rougher.output.recovery'], pred_test_rough,
                                target_test['final.output.recovery'],
                                pred_forest_final
                               )
```

```python
print('Итоговая оценка smape на тестовых данных равна', round(smape_test, 2))
```

И сравню с прогнозом на константной модели.

```python
def constant_model(features_train, target_train, features_test):

    dummy_regressor = DummyRegressor(strategy="mean")

    dummy_regressor.fit(features_train, target_train)

    prediction_test = dummy_regressor.predict(features_test)
    return prediction_test
```

Прогноз для этапа флотации.

```python
pred_test_rough_con= constant_model(features_train, target_train['rougher.output.recovery'], features_test)
```

Прогноз для финального концентрата.

```python
pred_test_fin_con = constant_model(features_train, target_train['final.output.recovery'], features_test)
```

```python
# рассчитаем smape на тестовых данных
smape_test_con = total_smape_function(target_test['rougher.output.recovery'],
                                pred_test_rough_con,
                                target_test['final.output.recovery'],
                                pred_test_fin_con
                               )
```

```python
print('Итоговая оценка smape константной модели на тестовых данных равна', round(smape_test_con, 3))
```
